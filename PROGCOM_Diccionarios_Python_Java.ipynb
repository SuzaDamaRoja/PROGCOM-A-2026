## 🧠 1. ¿Qué es un diccionario?

Un **diccionario** es una estructura de datos que guarda información en pares:

- **clave**: el identificador único
- **valor**: el dato asociado a esa clave

### Ejemplo de uso

Imagina que tu "inventario" de tu inventario se genera así:

- `"Pala": 15`
- `"Exp": 10000`
- `"Kinket": 12000`

Esto significa que, posees tres objetos:

- la clave `"Pala"` tiene como valor `15`
- la clave `"Exp"` tiene como valor `10000`
- la clave `"Kinket"` tiene como valor `12000`

Los diccionarios son útiles cuando quieres buscar datos rápidamente usando un nombre o identificador.

## 🐍 2. Diccionarios en Python

En Python, los diccionarios se escriben con llaves `{}`.

### Sintaxis general

```python
diccionario = {
    "clave1": valor1,
    "clave2": valor2
}
```

````python
# Inventario del minero
inventario = {
    "Pala": 15,
    "Exp": 10000,
    "Kinket": 12000
}

print(inventario)
````

### Operaciones básicas en Python

````python
inventario = {
    "Pala": 15,
    "Exp": 10000,
    "Kinket": 12000
}

# 1. Acceder a un valor
print("Daño de la pala:", inventario["Pala"])

# 2. Agregar un nuevo elemento
inventario["pocion"] = 3
print("Después de agregar poción:", inventario)

# 3. Modificar un valor existente
inventario["Kinket"] = 200
print("Después de modificar oro:", inventario)

# 4. Eliminar un elemento
del inventario["Exp"]
print("Después de eliminar escudo:", inventario)
````

## ☕ 3. Diccionarios en Java con `HashMap`

En Java, la estructura equivalente más común es `HashMap`.

### Sintaxis general

```java
import java.util.HashMap;

HashMap<String, Integer> inventario = new HashMap<>();
inventario.put("Pala", 15);
inventario.put("Exp", 10000);
inventario.put("Kinket", 12000);
```

### Operaciones básicas en Java

```java
import java.util.HashMap;

HashMap<String, Integer> inventario = new HashMap<>();

inventario.put("Pala", 15);
inventario.put("Exp", 10000);
inventario.put("Kinket", 12000);

// Acceder
System.out.println(inventario.get("Pala"));

// Agregar
inventario.put("pocion", 3);

// Modificar
inventario.put("Kinket", 200);

// Eliminar
inventario.remove("Exp");
```

### Idea clave

- En **Python** usamos `dict`
- En **Java** usamos con frecuencia `HashMap`
- En ambos casos la lógica es la misma: **clave → valor**

## ⚔️ 4. Ejercicio explicado 1 — Nivel muy sencillo


La archivista del reino guardó el poder de tres reliquias:

- espada: 20
- arco: 12
- bastón: 18

Se desea mostrar el poder de la espada.

---

### ✅ Solución en Python

````python
reliquias = {
    "espada": 20,
    "arco": 12,
    "bastón": 18
}

print("El poder de la espada es:", reliquias["espada"])
````

3. Para consultar un valor, se usa la clave entre corchetes:
   `reliquias["espada"]`
4. El resultado es `20`.

---

### ✅ Solución equivalente en Java

```java
import java.util.HashMap;

HashMap<String, Integer> reliquias = new HashMap<>();

reliquias.put("espada", 20);
reliquias.put("arco", 12);
reliquias.put("baston", 18);
```

## 🟡 5. Ejercicio explicado 2 — Complejidad media


Un guardián registra cuántas veces aparecen ciertos monstruos durante una expedición.

Lista observada:

`["slime", "dragon", "slime", "lobo", "dragon", "slime"]`

Se desea construir un diccionario que cuente cuántas veces aparece cada monstruo.

---

### ✅ Solución en Python

````python
monstruos = ["slime", "dragon", "slime", "lobo", "dragon", "slime"]
conteo = {}

for monstruo in monstruos:
    if monstruo in conteo:
        conteo[monstruo] += 1
    else:
        conteo[monstruo] = 1

print("Conteo final:", conteo)
````

### 🔍 Explicación paso a paso

1. Se parte de una lista de monstruos repetidos.
2. Se crea un diccionario vacío llamado `conteo`.
3. Se recorre la lista con un ciclo `for`.
4. En cada vuelta:
   - si el monstruo ya está en el diccionario, se suma 1
   - si no está, se crea con valor 1
5. Al final, el diccionario contiene la frecuencia de aparición.

### Resultado esperado

```python
{'slime': 3, 'dragon': 2, 'lobo': 1}
```

---

### ✅ Solución equivalente en Java

```java
import java.util.HashMap;

HashMap<String, Integer> conteo = new HashMap<>();
String[] monstruos = {"slime", "dragon", "slime", "lobo", "dragon", "slime"};

for (String monstruo : monstruos) {
    if (conteo.containsKey(monstruo)) {
        conteo.put(monstruo, conteo.get(monstruo) + 1);
    } else {
        conteo.put(monstruo, 1);
    }
}

System.out.println(conteo);
```

## 🧩 6. Resumen rápido: Python vs Java

| Aspecto | Python | Java |
|---|---|---|
| Estructura | `dict` | `HashMap` |
| Crear | `{}` | `new HashMap<>()` |
| Insertar | `dic["x"] = 5` | `map.put("x", 5)` |
| Consultar | `dic["x"]` | `map.get("x")` |
| Verificar existencia | `"x" in dic` | `map.containsKey("x")` |
| Eliminar | `del dic["x"]` | `map.remove("x")` |

### Idea central

Aunque cambie la sintaxis, en ambos lenguajes se trabaja con la misma lógica:

**usar una clave para encontrar un valor**

###Solución ejercicio 10
Como son varios distritos, voy a crear en cada distrito un diccionario. Mis distritos son:
* FUNPRO: Distrito minero
* PROGCOM: Distrito constructor
* SEMING: Distrito estratega

Como son inventarios, voy a crear un diccionario inventario, que va a contener el inventario de los otros distritos

````python
#Diccionario de diccionarios
inventarios = {
    "FUNPRO": {},
    "PROGCOM": {},
    "SEMING": {},
}
````

Ahora, voy a crear funciones que me permita agregar, actualizar, eliminar y mostrar el inventario final

La siguiente función agregar_objeto posee tres parámetros:
* distrito (string) que permite validar si hay una clave con ese nombre en inventarios. Cuáles claves hay en inventarios?
* nombre (string) que permite agregar el nuevo objeto en el inventario.
* cantidad (int) que permite agregar la cantidad de objetos a ingresar. Si El nombre no exxiste, lo crea. Si el nombre ya existía, agrega la cantidad

````python
def agregar_objeto(distrito, nombre, cantidad):
    if distrito in inventarios:
        if nombre in inventarios[distrito]:
            inventarios[distrito][nombre] += cantidad
        else:
            inventarios[distrito][nombre] = cantidad
    else:
        print("Este distrito no forma parte de Unabtopía.")
````

````python
#Ejemplo
agregar_objeto("FUNPRO", "Pala", 15)
````

Podrías explicar qué hace entonces las siguientes funciones?

````python
def actualizar_cantidad(distrito, nombre, nueva_cantidad):
    if distrito in inventarios and nombre in inventarios[distrito]:
        inventarios[distrito][nombre] = nueva_cantidad
    else:
        print("Objeto o distrito no existe.")


def eliminar_objeto(distrito, nombre):
    if distrito in inventarios and nombre in inventarios[distrito]:
        del inventarios[distrito][nombre]
    else:
        print("Objeto o distrito no existe.")


def mostrar_inventarios():
    print("\n=== INVENTARIOS DE UNABTOPIA ===")
    for distrito, inventario in inventarios.items():
        print(f"\n🏘️{distrito}:")
        if inventario:
            for objeto, cantidad in inventario.items():
                print(f" - {objeto}: {cantidad}")
        else:
            print(" (La bodega se encuentra actualmente vacía)")
````

Ahora debo crear un menú que le permita al usuario acceder a estos inventarios

````python
def menu():
  print("Ingrese un valor de acuerdo a lo que desee hacer:")
  print("1. Agregar un objeto")
  print("2. Actualizar la cantidad de un objeto")
  print("3. Eliminar un objeto")
  print("4. Mostrar el inventario final")
  print("5. Salir")
  opc=int(input("Ingrese su opción: "))
  return opc
````

Ahora, debo crear una función que dada una opción, llame a las siguientes funciones

````python
while True:
  opc=menu()
  if opc==1:
    distrito=input("Ingrese el distrito: ")
    nombre=input("Ingrese el nombre del objeto: ")
    cantidad=int(input("Ingrese la cantidad del objeto: \n"))
    agregar_objeto(distrito, nombre, cantidad)
  elif opc==2:
    distrito=input("Ingrese el distrito: ")
    nombre=input("Ingrese el nombre del objeto: ")
    nueva_cantidad=int(input("Ingrese la nueva cantidad del objeto: \n"))
    actualizar_cantidad(distrito, nombre, nueva_cantidad)
  elif opc==3:
    distrito=input("Ingrese el distrito: ")
    nombre=input("Ingrese el nombre del objeto: \n")
    eliminar_objeto(distrito, nombre)
  elif opc==4:
    mostrar_inventarios()
  elif opc==5:
    break
````
