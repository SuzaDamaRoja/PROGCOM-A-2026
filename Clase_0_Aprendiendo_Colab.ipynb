###Concatenación
Concatenar es unir dos tipos diferentes de datos

````python
x=int(input("Ingrese su edad: "))
y=input("Ingrese su nombre: ")
````

````python
#Ejemplo 1
print(y,"tiene una edad de",x)
````

````python
#Ejemplo 2
print(y+" tiene una edad de "+str(x))
````

````python
#Ejemplo 3
print(f"{y} tiene una edad de {x}")
````
